```python
student1=[]
student2=[]
count1 = 0
count2 = 0
count3 = 0
wintercamp = []
m=[]
n=[]
o=[]
p=[]
q=[]
r=[]
s=[]
t=[]
l=[]
X=df.universityRollNo
for i in range(2001):
    l.append(X[i])
for j in range(2001):
    if l[j]=='17BCS4566':  # Student 1 UID
        x=j
    elif l[j]=='17BCS4572': # Student 2 UID
        y=j
Y=df.Quantitative_Ability
for i in range(2001):
        m.append(Y[i])
        if i == x:
            student1.append(Y[i])
        elif i== y:
            student2.append(Y[i])
        elif Y[i]<500:
            count1 += 1
            wintercamp.append(count1)
Z=df.English_Comprehension
for i in range(2001):
        n.append(Z[i])
        if i == x:
            student1.append(Z[i])
        elif i== y:
            student2.append(Z[i])
        elif Z[i]<500:
            count2 += 1
            wintercamp.append(count2)
A=df.Logical_Ability
for i in range(2001):
        o.append(A[i])
        if i == x:
            student1.append(A[i])
        elif i== y:
            student2.append(A[i])
        elif A[i]<500:
            count3 += 1
            wintercamp.append(count3)
B=df.Computer_Programming
for i in range(2001):
        p.append(B[i])
        if i == x:
            student1.append(B[i])
        elif i== y:
            student2.append(B[i])
C=df.Computer_Science
for i in range(2001):
        q.append(C[i])
        if i == x:
            student1.append(C[i])
        elif i== y:
            student2.append(C[i])
D=df.Automata
for i in range(2001):
        r.append(D[i])
        if i == x:
            student1.append(D[i])
        elif i== y:
            student2.append(D[i])
E=df.Automata_Fix
for i in range(2001):
        s.append(E[i])
        if i == x:
            student1.append(E[i])
        elif i== y:
            student2.append(E[i])
F=df.WriteX_Essay_Writing_Total_Score
for i in range(2001):
        t.append(F[i])
        if i == x:
            student1.append(F[i])
        elif i== y:
            student2.append(F[i])

print(student1,end=' ')
print(student2)
```

## Compairing Scores between Two Students :-

```python
plt.plot(student1)
plt.plot(student2)
plt.legend(['STUDENT1','STUDENT2'])
plt.ylabel('Amcat SCORE')
plt.xlabel('Subjects')
plt.show()
```

```python
label = ['Quantitative', 'English', 'Logical', 'Computer programming','Computer Science','Automata','AutomataFix','Writex[5]']
marks = student1
marks2 = student2

def plot_bar_x():
    index = np.arange(len(label))
    plt.barh(index, marks)
    plt.barh(index, marks2)
    
    plt.xlabel('Subjects', fontsize=10)
    plt.ylabel('Scores', fontsize=10)
    plt.legend(['STUDENT1','STUDENT2'])
    plt.yticks(index, label, fontsize=10, rotation=30)
    plt.title('Subject wise scores STUDENT1-STUDENT2 AMCAT-2019')
    plt.show()
    
plot_bar_x()
```
